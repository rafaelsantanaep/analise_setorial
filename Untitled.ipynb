```python
def validating_and_converting(data, column):
    for index, values in enumerate(data[column]):
        if str(values).isdigit():
            data.loc[index, column] = values
        else:
            # Eu decidi representar os valores que estão faltando como -1s
            # já que o Python não aceita NaN em variáveis com o tipo integer
            data.loc[index, column] = -1

    return data[column].astype('int64')
```

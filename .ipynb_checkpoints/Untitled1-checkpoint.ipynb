## Introdução

Nesse projeto, será realizada uma análise sobre as empresas e sobre o número de empregados assaliarados no Brasil utilizando os dados do CNAE (Classificação Nacional de Atividades Econômicas), utilizando como espaço temporal os anos de 2014 e 2015.

Os dados já foram limpos através da utilização de um script do script `wrangling.py` que está presente nesse mesmo diretório e foram gerados um único arquivo que serão utilizados nessa análise. Foi possível realizar isso porque as quatro bases de dados possuem os mesmos identificadores: o identificador da seção e o identificador da divisão.

Após realizar todo o processo de limpeza, optei pela remoção de ambas colunas, já que elas não precisariam mais ser utilizadas nessa análise.


Alguns conceitos importantes precisão ser aqui trabalhados, com o intuito de melhorar o entendimento sobre os dados utilizados.

Na primeira base de dados, o IBGE identificou o número de empresas, funcionários assalariados e funcionários totais segmentando em dois: 
- Empresas com 10 funcionários ou mais que tem até oito anos de idade
- Todas empresas com 10 funcionários ou mais


Na segunda base de dados, o IBGE realizou o mesmo processo, porém, desta vez, segmentando as empresas
em dois tipos:
- Empresas gazelas (Empresas de alto crescimento que tem até oito anos de idade)
- empresas de alto crescimento

Dessa forma, vale notar que em ambas bases de dados, existe uma variável que é úm subconjunto da outra. Por exemplo, o número de funcionários das empresas gazelas, são um subconjunto do número de funcionários das empresas de alto crescimento, podendo, dessa forma, se utilizar de proporções para fazer comparações entre setores e divisões.

- Qual a parcela das empresas com até oito anos é representada pelas empresas gazelas.
- Qual a parcela das empresas de alto crescimento possui menos de oito anos?

Além disso, também será possível identificar entre as empresas com menos de oito anos de idade, quantas delas são empresas de alto crescimento (gazelas).


```r
library(dplyr)
library(ggplot2)
library(tidyr)
library(data.table)
library(repr)
library(ggplot2)
library(ggthemes)
theme_set(theme_light())
library(ggpubr)
setwd('/home/rafael/Documents/the_hive/the_hive_test/data/')
library(cowplot)
options("scipen"=999, "digits"=4)
```

```r
df <- fread('merged.csv')
```

#### Distribuição de funcionários totais (Empresas com até oito anos x Empresas com mais de oito anos)

```r
options(repr.plot.width=10, repr.plot.height=10)

numero_empresas <- ggecdf(data=subset(df, df$base_de_dados == 'geral'), 
                          x="numero_de_empresas",linetype='dashed', size=1,
                          color='segmento',
                          facet.by='year', title='Distribuição cumulativa de empresas por ano')

numero_funcionários <- ggecdf(data=subset(df, df$base_de_dados == 'geral'),
                          x='pessoal_ocupado_total', color='segmento',
                          linetype='dashed', size=1, facet.by='year', 
                          title='Distribuição cumulativa do numero de funcionários por tipo de empresa e por ano')

summary <- df %>% 
filter(base_de_dados == 'geral') %>% 
group_by(segmento) %>% 
summarize(mean_numero_de_empresas = mean(numero_de_empresas),
         mean_funcionários = mean(pessoal_ocupado_total),
         median_over_mean_empresas = median(numero_de_empresas) / mean(numero_de_empresas),
         median_over_mean_funcionários = median(pessoal_ocupado_total) / mean(pessoal_ocupado_total))

print(plot_grid(numero_empresas, numero_funcionários, nrow=2))
```

```r
print("Summary")
print(summary)
```

Como pode ser observado nesse ECDF, tanto as distribuições de funcionários quanto de número de empresas por divisão são enviesadas para a direita (skewed), dessa forma, possuem uma mediana menor que a média. Sendo importante destacar que isso não sofre muitas alterações quando se leva em conta o ano.

Entre as distribuições, a distribuição de funcionários é menos enviesada para a direita do que a distribuição de empresas, principalmente, quando se compara os valores relativos a todas as empresas com mais de dez funcionários.




```r
options(repr.plot.width=14, repr.plot.height=14)

percent_plot <- function(df, segmento_value, title='', x='', y='') {

# Total de empresas em 2014 com mais de 10 funcionários, independente do setor.
total_empresas <- df %>%
filter(base_de_dados == 'geral', year == '2014', segmento == 'Geral') %>%
summarise(total_empresas = sum(numero_de_empresas))

    
# Criando um plot para identificar 
plot <- df %>%
filter(base_de_dados == 'geral', year == '2014', segmento == segmento_value) %>%
mutate(percentual_total = numero_de_empresas / as.integer(total_empresas)) %>%
arrange(desc(percentual_total)) %>%
head(10) %>%
ggplot(aes(reorder(divisão_descrição, percentual_total),  100 * percentual_total, 
           fill=seção_descrição)) + geom_bar(stat='identity') +
scale_fill_brewer(type='qual') + 
coord_flip() +
geom_text(size=5, aes(label=numero_de_empresas), position = position_stack(vjust=0.5)) + 
labs(title=title, x=x, y=y, fill='número_de_empresas')

return(plot)
}

plot1 <- percent_plot(df, 'Geral', '10 divisões com maior parcela no total de empresas (Geral)')
plot2 <- percent_plot(df, 'Até 8 anos de idade', 
                      '10 divisões com maior parcela no total de empresas (Até oito anos)',
                      x='divisão', y='percentual do total de empresas')
plot_grid(plot1, plot2, nrow=2)
```

```r
options(repr.plot.width=15, repr.plot.height=7)

df %>%
filter(base_de_dados == 'geral') %>%
group_by(seção_descrição, segmento, year) %>%
summarize(numero_de_empresas = sum(numero_de_empresas)) %>%
spread(segmento, numero_de_empresas) %>%
mutate(perc = `Até 8 anos de idade` / Geral) %>%
select(c('seção_descrição','year','Até 8 anos de idade','perc')) %>%
drop_na() %>%
ggplot(aes(seção_descrição, perc * 100, fill=`Até 8 anos de idade`, label=`Até 8 anos de idade`)) + 
geom_bar(stat='identity', color='black') +
geom_text(size=4, position = position_stack(vjust=0.5)) +
facet_wrap(~year) + coord_flip() +
scale_fill_gradient2() +
labs(title='Parcela representada pelas empresas com até oito anos por setor',
    x='setor', y='% do total de empresas ', fill='numero de empresas')
```

```r
options(repr.plot.width=15, repr.plot.height=7)

df %>%
filter(base_de_dados == 'altocrescimento') %>%
group_by(seção_descrição, segmento, year) %>%
summarize(numero_de_empresas = sum(numero_de_empresas)) %>%
spread(segmento, numero_de_empresas) %>%
mutate(perc = Gazela / `Alto Crescimento`) %>%
select(c('seção_descrição','year','Gazela','perc')) %>%
drop_na() %>%
ggplot(aes(seção_descrição, perc * 100, fill=`Gazela`, label=`Gazela`)) + 
geom_bar(stat='identity', color='black') +
geom_text(size=4, position = position_stack(vjust=0.5)) +
facet_wrap(~year) + coord_flip() +
scale_fill_gradient2() +
labs(title='Parcela representada pelas empresas com até oito anos por setor',
    x='setor', y='% do total de empresas 10 funcionários ou mais', fill='numero de empresas')
```
